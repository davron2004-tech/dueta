```python
from datasets import load_dataset
from matplotlib import pyplot as plt
from tqdm import tqdm

def plot_token_length_distribution(dataset, token_limit):
    lengths = dataset["length"]

    lengths = [l for l in tqdm(lengths, desc="Filtering token lengths") if l <= token_limit]
    print(f"Number of token lengths below {token_limit}:", len(lengths))
    plt.hist(lengths, bins=50)
    plt.xlabel("Length")
    plt.ylabel("Frequency")
    plt.title(f"Distribution of Token Lengths below {token_limit}")
    plt.show()

dataset = load_dataset("davron04/dueta-tokenized", split="train")

plot_token_length_distribution(dataset, 128)
plot_token_length_distribution(dataset, 256)
plot_token_length_distribution(dataset, 512)
plot_token_length_distribution(dataset, 1024)
```
